# 輸出所有留言為Excel檔，做label

```python
file_name = "good.txt"
good= open(file_name,'r',encoding='utf-8').read()
good =list(set(good.split()))
    
file_name = "bad.txt"
bad= open(file_name,'r',encoding='utf-8').read()
bad =list(set(bad.split()))        

ask = ['嗎', '麼', '?', '？']
```

```python
import re

comment_orgin = df['留言']
comment_list, good_list, bad_list, ask_list = [], [], [], []

for comment in comment_orgin:
  good_count = 0
  bad_count = 0
  ask_count = 0
  if type(comment) != float:
    comment = comment[1:-1].split(',')

    for i in comment:
      comment_list.append(i[1:-1])

      for g in good:
        if g in i:
          good_count+=1
        else:
          continue

      for b in bad:
        if b in i:
          bad_count+=1
        else:
          continue

      for a in ask:
        if a in i:
          ask_count+=1
        else:
          continue
  good_list.append(good_count)
  bad_list.append(bad_count)
  ask_list.append(ask_count)


```

```python
df['好評留言'] = good_list
df['負評留言'] = bad_list
df['問句留言'] = ask_list
```
