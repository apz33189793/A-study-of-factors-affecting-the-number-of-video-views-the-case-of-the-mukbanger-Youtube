## Load and Clean Data 

```python
import pandas as pd
raw = pd.read_csv('yt_data.csv')
raw['len_title'] = [len(i) for i in raw['video_title']]
# raw.columnsM

d = raw['published_date']
y_lst = []
m_lst = []
d_lst = []
for w in d:
    splt = w.split('-', 3)
#     print(w.split('-', 3)[0])
    y_lst.append(splt[0])
    m_lst.append(splt[1])
    d_lst.append(splt[2])

raw['published_year'] = y_lst
raw['published_month'] = m_lst
raw['published_day'] = d_lst
raw['published_year'] = raw['published_year'].astype(int)
raw['published_month'] = raw['published_month'].astype(int)
raw['published_day'] = raw['published_day'].astype(int)

## normalize
raw_norm_col = ['view','commentCount','pos_comments', 'neg_comments', 'q_comments', 'len_title']
df_to_norm = raw[raw_norm_col]
# df_to_norm
df_normed = (df_to_norm-df_to_norm.min())/(df_to_norm.max()-df_to_norm.min())
# df_normed
other_var = ['channel_name', 'video_title', 'nano_ytb', 'type_cook', 'type_eat', 'type_unbox', 'type_lottery', 'type_car', 'type_life', 'type_travel']
df_str_cate = raw[other_var]
df_full = pd.concat([df_normed, df_str_cate], axis=1)
df_full.head(1)

df_using = df_full.drop(columns=['channel_name', 'video_title'])
df_big_ytb = df_using[df_using['nano_ytb']==0]
df_small_ytb = df_using[df_using['nano_ytb']==1]
df_small_ytb.head(1)
```

## Split Data

```python
import sklearn
from sklearn.model_selection import train_test_split
def double_split(X, y):
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
    X_subtrain, X_valid, y_subtrain, y_valid = train_test_split(X_train, y_train, test_size=0.2, random_state=42)
    return X_train, y_train, X_subtrain, y_subtrain, X_valid, y_valid, X_test, y_test

X_train, y_train, X_subtrain, y_subtrain, X_valid, y_valid, X_test, y_test = double_split(x_all, y_all)
```

## Define Prediction Models

```python
from sklearn.metrics import mean_squared_error
def rmse_func(y_pred, y_test):
    return mean_squared_error(y_pred, y_test, squared=False)
```

```python
from sklearn.model_selection import GridSearchCV
def model_tuning(model, param_grid_input, x_valid, y_valid, scoring_type):
    gs = GridSearchCV(model, param_grid = param_grid_input, scoring=scoring_type)
    gs.fit(x_valid, y_valid)
    best_params = gs.best_params_
    return best_params
```

```python
def model_train_rmse(model, x_train, x_test, y_train, y_test):
    model.fit(x_train, y_train)
    y_pred = model.predict(x_test)
    rmse_val = rmse_func(y_pred, y_test)
    return rmse_val
```

```python
def model_train_rmse_feat(model, x_train, x_test, y_train, y_test):
    model.fit(x_train, y_train)
    y_pred = model.predict(x_test)
    rmse_val = rmse_func(y_pred, y_test)
    
    feat_value = model.feature_importances_
    idx = np.flip(np.argsort(feat_value))
    top_feat_value = sorted(feat_value, reverse=True)[:5]
    top_feat_name = x_train.columns[idx[:5]]
    return rmse_val, top_feat_value, top_feat_name
```

#### 2. Gradient Boosting

```python
from sklearn.ensemble import GradientBoostingRegressor

print('=========== GradientBoosting ================')
model_gb = GradientBoostingRegressor(random_state=42)
param_grid_input = { 'loss':["squared_error", "huber", "absolute_error", "quantile"],
                   'learning_rate':np.linspace(0.01, 0.1, 10), 
                   'n_estimators':range(80, 120, 10)}
best_param = model_tuning(model_gb, param_grid_input, X_valid, y_valid, 'neg_root_mean_squared_error')
print('best_param:', best_param['n_estimators'], best_param['learning_rate'], best_param['loss']) #, best_param['loss'], best_param['learning_rate'])

best_model_gb = GradientBoostingRegressor(n_estimators=best_param['n_estimators'], random_state=42,
                                          loss=best_param['loss'], 
                                          learning_rate=best_param['learning_rate'])
gb_rmse_val, gb_top_feat_value, gb_top_feat_name = model_train_rmse_feat(best_model_gb, X_train, X_test, y_train, y_test)
print(gb_rmse_val)
print(gb_top_feat_value)
print(gb_top_feat_name)
```

#### 3. Decision Tree

```python
from sklearn.tree import DecisionTreeRegressor

model_dt = DecisionTreeRegressor(random_state=42)
param_grid_input = {'criterion':["squared_error", "friedman_mse", "absolute_error", "poisson"],
                   'max_depth':range(2, 10),
                   'min_samples_split':range(1, 10)}
best_param = model_tuning(model_dt, param_grid_input, X_valid, y_valid, 'neg_root_mean_squared_error')
print('best_param:', best_param['max_depth'], best_param['min_samples_split'], best_param['criterion'])

best_model_dt = DecisionTreeRegressor(max_depth=best_param['max_depth'], min_samples_split=best_param['min_samples_split'], 
                                      criterion=best_param['criterion'], random_state=42)
dt_rmse_val, dt_top_feat_value, dt_top_feat_name = model_train_rmse_feat(best_model_dt, X_train, X_test, y_train, y_test)
print(dt_rmse_val)
print(dt_top_feat_value)
print(dt_top_feat_name)
```

#### 4. Bagging

```python
from sklearn.ensemble import BaggingRegressor, RandomForestRegressor

model_bag = BaggingRegressor(random_state=42)
param_grid_inputs = {'n_estimators':range(80, 100, 5)}
gd = GridSearchCV(model_bag, param_grid_inputs)
gd.fit(X_valid, y_valid)
n_estimators = gd.best_params_['n_estimators']
print("best n_estimators:", n_estimators)

best_bagging = BaggingRegressor(n_estimators=n_estimators,
                               oob_score=True,
                               random_state=42)
best_bagging.fit(X_valid, y_valid)
pred = best_bagging.predict(X_test)
RMSE = mean_squared_error(y_test, pred, squared=False)
print('oob score:', best_bagging.oob_score_)
print('RMSE:', RMSE)
```

#### 5. Stacking

```python
from sklearn.ensemble import StackingRegressor
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import mean_squared_error

ests = [
        ("Random Forest", best_model_rf),
        ("Gradient Boosting", best_model_gb),
        ("Bagging", best_bagging)
        ]

stacking = StackingRegressor(estimators=ests)

stacking.fit(X_train, y_train)

pred = stacking.predict(X_test)
stack_RMSE = mean_squared_error(y_test, pred, squared=False)
print(stack_RMSE)
```
